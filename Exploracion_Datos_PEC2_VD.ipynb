## Transformación agrupación de valores por categoría 

En estas transformaciones creo las variables de Total_$Tema. De esta manera recojo para cada tema que se trata en los acuerdos su total, esto nos permitirá comparar temas entre acuerdos o procesos y ver los que se tratan más. 

```python
df['Total_Groups']= df.iloc[:,range(df.columns.get_loc('GCh'),df.columns.get_loc('GeWom'))].sum(axis=1)
df['Total_Gender']= df.iloc[:,range(df.columns.get_loc('GeWom'),df.columns.get_loc('StDef'))].sum(axis=1)
df['Total_StateDef']= df.iloc[:,range(df.columns.get_loc('StDef'),df.columns.get_loc('Pol'))].sum(axis=1)
df['Total_Governance']= df.iloc[:,range(df.columns.get_loc('Pol'),df.columns.get_loc('Polps'))].sum(axis=1)
df['Total_Powersharing']= df.iloc[:,range(df.columns.get_loc('Polps'),df.columns.get_loc('HrGen'))].sum(axis=1)
df['Total_HR']= df.iloc[:,range(df.columns.get_loc('HrGen'),df.columns.get_loc('JusCr'))].sum(axis=1)
df['Total_Justice']= df.iloc[:,range(df.columns.get_loc('JusCr'),df.columns.get_loc('Dev'))].sum(axis=1)
df['Total_SocioEco']= df.iloc[:,range(df.columns.get_loc('Dev'),df.columns.get_loc('LaRef'))].sum(axis=1)
df['Total_Land']= df.iloc[:,range(df.columns.get_loc('LaRef'),df.columns.get_loc('SsrGua'))].sum(axis=1)
df['Total_Security']= df.iloc[:,range(df.columns.get_loc('SsrGua'),df.columns.get_loc('TjGen'))].sum(axis=1)
df['Total_Transition']= df.iloc[:,range(df.columns.get_loc('TjGen'),df.columns.get_loc('ImUN'))].sum(axis=1)
df['Total_Implementation']= df.iloc[:,range(df.columns.get_loc('ImUN'),df.columns.get_loc('ImSrc'))].sum(axis=1)
```

Estudio de los tipo de conflicto por proceso de paz

```python
from collections import Counter
import numpy as np
def fixOverLappingText(text):

    # if undetected overlaps reduce sigFigures to 1
    sigFigures = 4
    positions = [(round(item.get_position()[1],sigFigures), item) for item in text]

    overLapping = Counter((item[0] for item in positions))
    overLapping = [key for key, value in overLapping.items() if value >=0.001]

    for key in overLapping:
        textObjects = [text for position, text in positions if position == key]

        if textObjects:

            # If bigger font size scale will need increasing
            scale = 0.05

            spacings = np.linspace(0,scale*len(textObjects)*2,len(textObjects))

            for shift, textObject in zip(spacings,textObjects):
                textObject.set_y(key + shift)
```

```python
import math
i = 0
fig, axs = plt.subplots(5,2,figsize=(20,20))
for proceso in top10_procesos:
    datos=df.loc[df['PPName'] == proceso].Contp.value_counts()/df.loc[df['PPName'] == proceso].shape[0]
    eh = pd.DataFrame([0,0,0,0,0], index=['Government/territory', 'Government', 'Inter-group', 'Territory', 'Other'])
    eh[1]=(datos)
    eh=eh.fillna(0)
    labels = eh.index
    text = axs[i%5,math.floor(i/5)].pie(x=eh[1], autopct="%.1f%%", explode=[0.05]*len(eh[1]), labels=labels, pctdistance=0.5)[1]
    fixOverLappingText(text)
    axs[i%5,math.floor(i/5)].set_title(proceso, fontsize=14);
    i = i+1

plt.show()
```

Como esto solo es una exploración de los datos no me he centrado en que los gráficos estén limpios y perfectos, solo entendibles para la exploración.  
Vemos que aunque en teoría hay una distribución de acuerdos de paz por tipo de conflicto general definida, para cada proceso no se mantiene esta distribución, sino que cada proceso de paz tiene una distribución diferente de tipos de conflicto. Lo que confirma el pensamiento de que los acuerdos dentro de un proceso de paz son similares.

Vemos que la mayoría de procesos tienen la mayoría de acuerdos de tipo gobierno/territorio, pero otros como los de colombia son plenamente gobierno y otros como el de Siria son casi todo grupos locales.

```python
import math
i = 0
fig, axs = plt.subplots(5,2,figsize=(20,20))
for proceso in top10_procesos:
    datos=df.loc[df['PPName'] == proceso].Agtp.value_counts()/df.loc[df['PPName'] == proceso].shape[0]
    eh = pd.DataFrame([0,0,0,0], index=['Intra', 'IntraLocal', 'InterIntra', 'Inter'])
    eh[1]=(datos)
    eh=eh.fillna(0)
    labels = eh.index
    text = axs[i%5,math.floor(i/5)].pie(x=eh[1], autopct="%.1f%%", explode=[0.05]*len(eh[1]), labels=labels, pctdistance=0.5)[1]
    fixOverLappingText(text)
    axs[i%5,math.floor(i/5)].set_title(proceso, fontsize=14);
    i = i+1

plt.show()
```

Lo mismo ocurre con los tipos de acuerdo, de acuerdo a la generalidad, la mayoría de acuerdos son de tipo Intra, sin embargo, vemos que algunos procesos tienen casi todo acuerdos intra, otros tienen algunos de InterIntra con la mayoría de Intra y otros procesos como el de Siria tienen dominio de acuerdos IntraLocales.

En definitiva, esto y el aspecto anterior confirman que los procesos contienen acuerdos parecidos entre ellos, por lo menos en estos aspectos.

# Curiosidades

Algo interesante de estudiar tras ver que los acuerdos dentro de un proceso se parecen, es ver si hay relación entre que un acuerdo trate un tema y que trate otros temas más o menos. 

Lo primero que voy a comprobar es si hay correlación entre que cuando un tema se trata más otros también se tratan más en los acuerdos de paz de manera general. Como ejemplo, ver si cuando los acuerdos de paz tienden a tratar temas de grupos sociales, se trata también temas de género en sus acuerdos. 

Esto lo estudio con una matriz de correlación para todos los acuerdos de paz del dataset.

```python
totales= df[['PPName','Total_Groups','Total_Gender','Total_StateDef','Total_Governance','Total_Powersharing','Total_HR','Total_Justice','Total_SocioEco','Total_Land','Total_Security','Total_Transition','Total_Implementation']]
```

```python
plt.figure(figsize=(12,8))
sn.heatmap(totales.corr('spearman'), annot=True)
plt.show()
```

En la matriz de correlación/heatmap podemos ver como se cumple lo esperado, algunos temas de acuerdos tienen mayor correlación con otros. Vemos como en los acuerdos que se tratan los grupos, se tratan normalmente también los géneros, y los derechos humanos dando una sensación de que hay acuerdos muy sociales/humanísticos.  
Por otra parte vemos como hay correlación entre acuerdos de Governance con Powersharing y Governance con HR, pero hay muy poca correlación en los acuerdos entre HR y Powersharing, todo esto podría indicar que cuando se trata el tema de gobierno, se puede hacer desde una visión social, o desde una visión de control del poder total.  
También destaca que hay correlación entre los acuerdos con alto interés socioeconómico y aquellos con alto interés de tierra, indicando que estos pueden ser otro tipo de acuerdos que se dan en los procesos de paz. 


**Al ver que la hipótesis parece cumplirse, y que los acuerdos de paz suelen tratar algunos temas en concreto más que otros según el acuerdo, podemos estudiar si éntre los procesos de paz se mantiene esta correlación. Es decir, ver si a pesar de que en un proceso de paz tiene muchos acuerdos diferentes, hay procesos de paz que tienden a tratar más temas sociales, o más temas económicos o más temas de control del poder, entre otros.**

Como ya hemos visto, los acuerdos de los procesos de paz del top 10 son parecidos entre ellos, ya que cada proceso trata unos conflictos, por tanto, podemos esperar que vamos a obtener lo mismo que hasta ahora o más, y ver que los procesos que tratan unos temas tienden a tratar otros temas. 

```python
totales_por_PPName=totales.groupby("PPName").mean()
```

```python
plt.figure(figsize=(12,8))
sn.heatmap(totales_por_PPName.corr('spearman'), annot=True)
plt.show()
```

Vemos que las correlaciones de la media de los temas que se tratan en un acuerdo entre los diferentes procesos se mantienen, es decir vemos de nuevo que generalmente en un proceso cuando crece el tratamiento de aspectos de grupos sociales, crece el de género y el de derechos humanos principalmente. Vemos también el resto de correlaciones mencionadas hasta ahora.

**Encontrar estas correlaciones entre los temas tratados no aporta solo la curiosidad de que hay agrupaciones de temas en los procesos y acuerdos como por ejemplo los grupos con género, gobierno y derechos humanos, sino que también confirma lo que ya hemos visto hasta ahora, que cada proceso contiene acuerdos similares entre ellos, puesto que las correlaciones han aumentado, y que hay procesos que tratan temas diferentes a otros procesos.**

Si simplemente miramos los acuerdos de paz a lo largo de los años, podemos ver como en la decada de los 90 hubo muchos acuerdos de paz, estos probablemente debidos a la partición de Yugoslavia, que ya hemos visto que es el país con más acuerdos.

```python
plt.figure(figsize=(12,8))

sn.regplot(x=fecha.Año, y=fecha.Acuerdos, data=fecha)
plt.xticks(rotation=15)
plt.title('Tendencia de acuerdos de paz por año')
plt.show()
```
